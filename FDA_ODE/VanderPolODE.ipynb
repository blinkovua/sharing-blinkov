```python
def f(y, args):
    mu = args[0]
    return array([\
        y[1],\
        -mu*(y[0]**2 - 1)*y[1] - y[0],\
    ], dtype=float64)
```

## Van der Pol oscillator

```python
xk, h = 80., 0.05
x = arange(0., xk, h, dtype=float64)
res = []
for mu in ([0., 0.5, 1., 2., 5., 10., 20., 50.]):
    y = empty((len(x), 2), dtype=float64)
    y[0] = 0., 1.
    for i in range(1, len(x)):
        y[i] = y[i-1] + RK4(f, [mu], y[i-1], h)
        if abs(y[i][0]) >= 5.:
            break
    res.append((mu, i, y))
    print(mu, h**4*mu**5)
```

```python
fig, axes = subplots(figsize=(12, 9), nrows = 4, ncols = 2,\
        layout='constrained', sharex=True, sharey=True)
for k in range(4):
    for j in range(2):
        mu, i, y = res[2*k+j]
        axes[k, j].set_title(r"$h$=%.1e, $\mu$=%.1f, $h^4\mu^5$=%.1e, max=%.3f" %
                (h, mu, h**4*mu**5, max(abs(y[:i, 0]))))
        axes[k, j].plot(x[:i], y[:i, 0])
        axes[k, j].grid()
        axes[k, j].set_ylim([-2.5, 2.5])
show() if isShow else savefig('figVanderPolRK4.pdf')
close()
```
