```python
import json

from pylab import *
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

from itertools import cycle
lstyle = cycle(('-', '--', ':', '-.',\
    (0, (1, 10)), (0, (5, 10)),\
    (0, (5, 5)), (0, (3, 10, 1, 10)),\
    (0, (3, 5, 1, 5)), \
    (0, (3, 5, 1, 5, 1, 5))))
color = cycle(('#1f77b4', '#ff7f0e', '#2ca02c',\
    '#d62728', '#9467bd', '#8c564b', '#e377c2',\
    '#7f7f7f', '#bcbd22', '#17becf'))
```

```python
lst = ("FDA1", "FDA2", "FDA3", "FDA4", "MAC")
```

```python
res = []
for r in lst:
    with open('2D-%s.json' % r) as f:
        data = json.load(f)
        res.append((r, data))
```

```python
l = ['u', 'v', 'p', 'div']
def draw(h):
    for k in range(1, 5):
        lstyle = cycle(('-.', '--', ':', '--'))
        color = cycle('rbgcmy')
        figure(figsize=(9, 6))
        title("%s, h = %s" % (l[k-1], h), {'fontsize': 18})
        trn = int(10/float(h))
        for r in res:
            n, err = r[1][h]
            semilogy(err[0][trn:n],\
                     err[k][trn:n],\
                     label="%s" % r[0],
                     linestyle=next(lstyle),
                     color=next(color),
                     linewidth=4)
        xlabel(r'$t$', {'fontsize': 18})
        grid()
        legend(loc='lower right', fancybox=True)
        #     savefig("taylor_decaying_uvp_mac.pdf")
        show()
        close()
```

```python
draw("0.2")
```

```python
draw("0.1")
```

```python
draw("0.05")
```
